```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from scipy import stats
```

### Tworzę dataset 1 i 2 (100 sampli i 10000 sampli)

```python
mean=3
std_dev=1
num_samples1=100
num_samples2=10000
data1=np.random.normal(mean,std_dev,num_samples1)
data2=np.random.normal(mean,std_dev,num_samples2)
```

### Pierwszy dataset (100 sampli):

```python
mean_data1=np.round(np.mean(data1))
print('Średnia:')
print(mean_data1)

median_data1=np.median(data1)
print('Mediana:')
print(median_data1)

mode_data1=stats.mode(data1)
print('Moda:')
print(mode_data1)


q0=np.quantile(data1,0.0)
q1=np.round(np.quantile(data1,0.25))
q3=np.round(np.quantile(data1,0.75))
q4=np.round(np.quantile(data1,1.0))

print('Kwartyle:')
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_data1}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')

plt.hist(data1, bins=20)
plt.xlabel('Wartość zmiennych')
plt.ylabel('Ilość zmiennych')
plt.axvline(x=mean_data1, color='red', label='Średnia')
plt.axvline(x=median_data1, color='green', label='Mediana')
plt.axvline(x=mode_data1[0], color='orange', label='Moda')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()

plt.boxplot(data1)
plt.show()
```

```python
range_data1=max(data1)-min(data1)
print('Zakres:')
print(range_data1)

iqr=q3-q1
print('Rozstęp międzykwartylowy:')
print(iqr)

variance_data1=np.var(data1, ddof=1)
print('Wariancja:')
print(variance_data1)

standard_deviation_data1=np.std(data1, ddof=1)
print('Odchylenie Standardowe:')
print(standard_deviation_data1)
```

### W drugim przypadku mamy do czynienia z większym zakresem wartości (pierwszy ok.5, drugi prawie 8)
